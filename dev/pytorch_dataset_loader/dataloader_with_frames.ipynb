```python
from torch.utils.data import DataLoader
from flytracker.io import VideoDataset
from flytracker.preprocessing import preprocessing
import numpy as np
import matplotlib.pyplot as plt
from flytracker import run

%load_ext autoreload
%autoreload 2
```

# Making mask

```python
movie_path = "../../data/experiments/clement/videos/seq_1.mp4"
mapping_folder = "../../data/distortion_maps/"
```

```python
# Making mask
mask = np.ones((1080, 1280), dtype=bool)
mask[:170, :] = 0
mask[-155:, :] = 0
mask[:, :240] = 0
mask[:, -275:] = 0
```

```python
dataset = VideoDataset(movie_path, preprocessing, mask, mapping_folder)
loader = DataLoader(dataset, batch_size=1, pin_memory=True)
```

```python
dataset.set_frame(100)
```

```python
next(enumerate(loader))[1]
```

```python
print("Running tracker.")
df = run(movie_path, mask, n_arenas=4, mapping_folder=mapping_folder, gpu=True, n_frames=1000)
```
