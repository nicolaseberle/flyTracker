```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_hdf('df.hdf', key='df')
```

```python
ID = 0
plt.plot(*df.query(f"ID == {ID}")[["x", "y"]].to_numpy().T, 'o')
```

```python
for ID in df.ID.unique():
    plt.plot(*df.query(f"ID == {ID}")[["x", "y"]].to_numpy().T, '-')
```

```python
arena = 3
for ID in df.query(f'arena == {arena}').ID.unique():
    plt.plot(*df.query(f"ID == {ID}")[["x", "y"]].to_numpy().T, '-')
```
